Boxplot, juga dikenal sebagai diagram kotak, adalah salah satu jenis visualisasi yang digunakan untuk menganalisis dan memvisualisasikan distribusi data. Boxplot memberikan gambaran ringkas tentang posisi pusat data, penyebaran data, serta adanya pencilan (outliers) dalam kumpulan data yang diamati. Metode ini dikembangkan oleh John Tukey pada tahun 1970.

Berikut adalah komponen-komponen utama yang terdapat dalam sebuah boxplot:

1. Median (Tengah): Garis tengah dalam kotak (box) menunjukkan nilai median data. Median adalah nilai tengah yang membagi data menjadi dua bagian setara.

2. Quartile (Kuartil): Kotak (box) pada boxplot dibentuk oleh kuartil pertama (Q1) dan kuartil ketiga (Q3). Kuartil pertama (Q1) adalah nilai yang membagi data menjadi dua bagian, di mana 25% data berada di bawahnya. Kuartil ketiga (Q3) adalah nilai yang membagi data menjadi dua bagian, di mana 75% data berada di bawahnya.

3. Rentang antarkuartil (Interquartile Range/IQR): Rentang antarkuartil didefinisikan sebagai selisih antara kuartil ketiga (Q3) dan kuartil pertama (Q1). Rentang ini memberikan informasi tentang penyebaran data pada bagian tengah distribusi.

4. Whisker (Janggut): Garis yang menghubungkan kotak (box) dengan garis-garis yang mengarah ke atas (atas) dan ke bawah (bawah) disebut janggut (whisker). Janggut atas mencapai batas data yang tidak tergolong sebagai pencilan (outlier) dan janggut bawah mencapai batas data yang tidak tergolong sebagai pencilan (outlier).

5. Pencilan (Outlier): Pencilan adalah data yang berada jauh dari nilai-nilai lain dalam kumpulan data. Pencilan ditandai sebagai titik-titik di luar janggut (whisker) pada boxplot.

Boxplot dapat memberikan pemahaman yang jelas tentang distribusi data dan memungkinkan kita untuk melihat adanya pencilan atau anomali dalam data. Diagram ini juga membantu dalam perbandingan distribusi data antara beberapa kelompok atau kategori yang berbeda. Dengan menganalisis boxplot, kita dapat menggambarkan bagaimana data terbagi dan menentukan apakah terdapat kehadiran nilai ekstrem atau anomali dalam kumpulan data.

Dalam boxplot, distribusi data direpresentasikan oleh kotak yang berada di antara kuartil pertama dan ketiga, dengan garis di dalam kotak yang menunjukkan median. Garis di atas dan di bawah kotak (yang disebut kumis) menandakan rentang data yang tidak dianggap sebagai pencilan. Pencilan yang dianggap signifikan dapat ditampilkan sebagai titik-titik individu di luar garis kumis.

Boxplot memberikan gambaran visual yang jelas tentang bagaimana data terdistribusi, termasuk adanya pencilan dan nilai-nilai ekstrem. Hal ini memudahkan pemahaman terhadap perbedaan antara kelompok data, serta mengidentifikasi pola dan karakteristik yang mungkin ada dalam data tersebut. Boxplot sering digunakan dalam analisis statistik, eksplorasi data, dan pembandingan kelompok.
